```python
import networkx as nx
from cdlib import algorithms, viz
import matplotlib.pyplot as plt
import igraph as ig
```

```python
g = nx.karate_club_graph()
g = ig.Graph.from_networkx(g)
```

```python
g.summary()
```

### Louvain

```python
coms = g.community_multilevel()
```

```python
ig.plot(coms, mark_groups=True)
```

### Fast Greedy

```python
coms = g.community_fastgreedy()
```

```python
ig.plot(coms, mark_groups=True)
```

### Spin glass

```python
coms = g.community_spinglass()
```

```python
ig.plot(coms, mark_groups=True)
```

### Walktrap

```python
coms = g.community_walktrap()
```

```python
ig.plot(coms, mark_groups=True)
```

### Edge betweenness

```python
coms = g.community_edge_betweenness()
```

```python
ig.plot(coms)
```

### Infomap

```python
coms = g.community_infomap()
```

```python
ig.plot(coms, mark_groups=True)
```

### Label propagation

```python
coms = g.community_label_propagation()
```

```python
ig.plot(coms, mark_groups=True)
```
